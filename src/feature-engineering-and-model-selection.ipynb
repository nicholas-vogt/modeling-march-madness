# Feature Engineering

Let's get started with our DataFrame. For simplicity's sake, we're only going to look at select data.

* FG: Field Goals Made
* FGA: Field Goals Attempted
* FT: Free Throws Made
* FTA: Free Throws Attempted
* PTS: Points Scored
* PF: Personal Fouls

And we're going to look at the Michigan State Spartans season in particular.

````python
variables = ['Date', 'TeamName', 'FG', 'FGA', 'FT', 'FTA', 'ScoreDiff', 'PF']
GameStats = get_season_game_stats()[variables]
MSUStats = GameStats[GameStats['TeamName'] == 'Michigan State']
# # Note that we've already sorted by date.
MSUStats = MSUStats.set_index(['Date', 'TeamName'])
MSUStats.head()
````

### Lag Variables

Lag (and lead) variables are commonly used in time series analysis. They're mostly self-explanatory. A 1-game lag on the PF (Personal Fouls) variable would return the PF value of the previous game. 

````python
# leads are done with negative integers
Lags = MSUStats.groupby(level=1).shift(periods=1)  
Lags.columns = [c+'Lag1' for c in Lags.columns]
Lags.head()
````

### Cumulative Summations

Cumulative sums have obvious applications.

````python
CumSums = Lags.groupby(level=1).cumsum()
CumSums.columns = ['Cum'+c[:-4] for c in CumSums.columns]
CumSums.head()
````

### Rolling Summations

We may care more about the trend instead of the most recent game. In this case, we'll do a rolling mean.

Note that there is no specific DataFrame method for this operation. Instead we use `apply`. It's easy enough to understand.

````python
RollingSums = Lags.groupby(level=1).apply(pd.rolling_mean, window=3, min_periods=1)
RollingSums.columns = [c[:-4]+'3GameAvg' for c in RollingSums.columns]
RollingSums.head()
````

# Preparing the Model

Let's prep the data for our logit model.

````python
Boxscores = get_boxscores()
# Sum all players while preserving date and teamname variables
GameStats = Boxscores.set_index(['Date', 'TeamName']).groupby(level=[0,1]).sum().reset_index()
GameResults = get_game_results()[['Date', 'TeamName', 'ScoreDiff']]
GameResults['Win'] = GameResults['ScoreDiff'] > 0
GameStats = pd.merge(GameResults, GameStats, on=['TeamName', 'Date'])

GameStats = GameStats.set_index(['Date', 'TeamName'])
variables = ['FG', 'FGA', 'FT', 'FTA', 'ScoreDiff', 'PF', 'Win']
GameStats = GameStats[variables].groupby(level=[0,1]).sum()
GameStats.head()

# Derive Lag and Rolling Mean Variables
Lags = GameStats.groupby(level=1).shift(periods=1)  
Lags.columns = [c+'Lag1' for c in Lags.columns]
RollingMeans = Lags.groupby(level=[1]).apply(pd.rolling_mean, window=3, min_periods=1)
RollingMeans.columns = [c[:-4]+'3GameAvg' for c in RollingMeans.columns]

Teams = put_team_ids()
Games = put_game_ids()
# Merge all variables
GameStats = GameStats.merge(Lags, left_index=True, right_index=True) \
           .merge(RollingMeans, left_index=True, right_index=True) \
           .reset_index() \
           .merge(Teams, on='TeamName') \
           .merge(Games, on=['Date', 'TeamID']) \
           .reset_index() 

del GameStats['Date']
del GameStats['index']
GameStats.tail()
````

````python
# Prep Data
Games = put_game_ids()[['GameID', 'TeamID']]

Team1GameStats = Games.groupby('GameID').first() \
                 .reset_index() \
                 .merge(GameStats, on=['GameID', 'TeamID'])
Team2GameStats = Games[::-1].groupby('GameID').first() \
                       .reset_index() \
                       .merge(GameStats, on=['GameID', 'TeamID'])
Team2GameStats.columns = ['Opp'+c if c not in ('GameID', 'TeamID') else c for c in Team2GameStats.columns]
Data = pd.merge(Team1GameStats, Team2GameStats, on='GameID')
# Data.to_csv('./../data/deletable.csv')
Data.head()
````

# Running the Model

````python
import statsmodels.api as sm

y_vars = 'Win'
X_vars = ['FG3GameAvg', 'FGA3GameAvg', 'FT3GameAvg',
          'FTA3GameAvg', 'PF3GameAvg', 'ScoreDiff3GameAvg', 'Win3GameAvg',
          'OppFG3GameAvg', 'OppFGA3GameAvg', 'OppFT3GameAvg',
          'OppFTA3GameAvg', 'OppPF3GameAvg', 'OppScoreDiff3GameAvg', 'OppWin3GameAvg']


Data = Data[[y_vars] + X_vars]
Data = Data.dropna()
y = Data[y_vars]
X = Data[X_vars]
X = sm.add_constant(X)

# X = Data[[c for c in Data.columns if c != 'Win']]
logit = sm.Logit(y, X)
print(logit.fit().summary())
````

````python
# print(logit.fit().summary())
Data['Pred'] = logit.fit().predict(X)
Data['PredWin'] = np.round(Data['Pred'])
print('accuracy =', np.mean(Data['PredWin'] == Data['Win']))
````
